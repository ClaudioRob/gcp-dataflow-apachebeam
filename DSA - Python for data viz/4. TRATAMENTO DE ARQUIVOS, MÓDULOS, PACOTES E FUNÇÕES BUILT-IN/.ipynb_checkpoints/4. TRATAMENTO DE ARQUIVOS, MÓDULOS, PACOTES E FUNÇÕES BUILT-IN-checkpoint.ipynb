### Dividindo um dataset em colunas

```python
# repete processo de abertura, leitura e separação de linhas por enter
f = open('arquivos/salarios.csv', 'r')
data = f.read()
rows = data.split('\n')
```

```python
# Cria uma lista vazia
full_data = []
```

```python
# loop for: para cada linha em tabela rows, cria variável split_row que 
# separa row por ",", depois grava split_row com append em lista full_data
for row in rows:
    split_row = row.split(",")
    full_data.append(split_row)
print(full_data)
```

### Contando as linhas de um arquivo

```python
# repete o processo anterior
# depois cria variável count = 0, e loop para row no set full_data
# conto 1 e salvo ao valor de count
count = 0
for row in full_data:
    count += 1   # Equivalente a: count = count + 1
    
print(count)
```

### Contando o número de colunas de um arquivo

```python
# antes, preciso rodar a criação de rows (rows = data.split('\n')), que é
# a tabela linha a linha
for row in rows:
    split_row = row.split(",")
    full_data.append(split_row)
    first_row = full_data[0]
count = 0

for column in first_row:
    count = count + 1
    
# Outra solução possível
# for column in full_data[0]:
#     count = count + 1

print(count)
```

### Manipulando Arquivos TXT

```python
texto = "Texto exemplo com um monte de caracteres."
```

```python
texto += "Esse é um segundo texto, só pra mostrar como concatena."
```

```python
print(texto)
```

Para trabalhar com texto, usa-se normalmente o pacote OS, que vamos chamar através do import, logo abaixo.

```python
# Importando o módulo os
import os 
```

```python
# criando um arquivo. No momento de criar, se não existe o arquivo, o python cria
arquivo = open (os.path.join('arquivos/cientista.txt'),'w')
```

```python
# Gravando os dados no arquivo. Para cada palavra in texto, usa módulo split
# escreve na variável arquivo, cada palavra e concatena com espaço
for palavra in texto.split():
    arquivo.write(palavra+' ')
```

```python
# Fechando o arquivo
arquivo.close()
```

```python
# Lendo o arquivo
arquivo = open('arquivos/cientista.txt','r')
conteudo = arquivo.read()
arquivo.close()

print(conteudo)
```

### Usando a expressão `with` 

O método close() é executado automaticamente

```python
with open('arquivos/cientista.txt','r') as arquivo:
    conteudo = arquivo.read()
```

```python
print(len(conteudo))
```

```python
# usando open para escrever dentro do arquivo, inserindo a variavel texto
# que está fatiada com slicing
with open('arquivos/cientista.txt','w') as arquivo:
    arquivo.write(texto[:21])
    arquivo.write('\n')
    arquivo.write(texto[:33])
```

```python
# Lendo o arquivo
arquivo = open('arquivos/cientista.txt','r')
conteudo = arquivo.read()
arquivo.close()

print (conteudo)
```
